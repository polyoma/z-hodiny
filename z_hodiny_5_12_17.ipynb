```python
import pandas as pd
import numpy as np
```

```python
telefon = pd.read_csv('https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/datasets/WorldPhones.csv')
telefon.head()
```

```python
telefon_melt = pd.melt(
    telefon, id_vars=['Unnamed: 0'], 
    value_vars=['N.Amer', 'Europe','Asia','S.Amer','Oceania', 'Africa', 'Mid.Amer'])
telefon_melt.head()
```

```python
telefon_melt
```

```python
telefon_melt.rename(columns={'Unnamed: 0':'year'}, inplace=True)
telefon_melt
```
